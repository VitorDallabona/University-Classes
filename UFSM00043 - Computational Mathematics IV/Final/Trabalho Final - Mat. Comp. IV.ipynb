# Difusão Magnética na Magnetoidrodinâmica (MHD)

A MHD relaciona as equações de Maxwell com a Hidrodinâmica clássica (equações de Navier-Stokes) para descrever o comportamento macroscópico de fluídos supercondutores, como plasma (estado da matéria composto por gás, totalmente ou parcialmente, ionizado).

Essa teoria é essencial na astrofísica (ex.: dinâmica do Sol, difusão de campos magnéticos em nuvens interestelares) e em aplicações de engenharia.

Neste notebook, será explorado a equação de **difusão magnética**, que modela o decaimento de campos magnéticos devido à resistividade do meio.

A MHD nas condições ideiais envolvem várias aproximações, a saber:

- velocidade de fluxo é bastante não relativística;
- condutividade elétrica é tão alta que pode ser considerada como infinita;
- baixa frequência, longo comprimento de onda.

## As equações de Difusão Magnética

As equações de difusão são obtidas essencialmente pela manipulação das equações de Maxwell (antes de sua forma final) e pela lei de Ohm. Consideremos as equações de pré-Maxwell (antes da unificação e correções por Maxwell) na forma diferencial:

- **Lei de Faraday**: Descreve como um campo magnético variável no tempo induz um campo elétrico.
$$
\vec{\nabla} \times \vec{E} = - \frac{\partial \vec{B}}{\partial t}
$$

- **Lei de Gauss**: Descreve como as cargas elétricas criam campos elétricos.
$$
\vec{\nabla} \cdot \vec{E} = \frac{\rho}{\varepsilon_0}
$$

- **Lei de Ampère**: Relacionava o campo magnético a correntes elétricas estacionárias (constantes no tempo).
$$
\vec{\nabla}\times \vec{B} = \mu_0 \vec{J}
$$

- **Lei de Gauss do magnetismo**: Estabelece a inexistência de monopolos magnéticos, ou seja, "cargas magnéticas" isoladas. 
$$
\vec{\nabla}\cdot \vec{B} = 0
$$

Se substituirmos a **Lei de Ohm** $(\vec{J} = \sigma \vec{E})$ na Lei de Ampère, obtemos:

$$
\vec{\nabla} \times \vec{B} = \mu_0 \sigma \vec{E}
$$

Tomando o rotacional dessa equação

$$
\vec{\nabla} \times \vec{\nabla}\times \vec{B} = \mu_0 \sigma \vec{\nabla}\times \vec{E}
$$

Pela identidade vetorial $\vec{\nabla}\times \vec{\nabla} \times \vec{A} = \vec{\nabla} (\vec{\nabla}\cdot \vec{A}) - \vec{\nabla}^2 \vec{A})$ obtemos

$$
\vec{\nabla}(\vec{\nabla}\cdot \vec{B}) - \vec{\nabla}^2 \vec{B} = -\mu_0 \sigma \frac{\partial \vec{B}}{\partial t}
$$

Como pela lei de Gauss do magnetismo temos $\vec{\nabla}\cdot \vec{B}$, obtemos:

$$
\boxed{\vec{\nabla}^2 \vec{B} = \mu_0 \sigma \frac{\partial \vec{B}}{\partial t}}
$$

De modo similar, tomando o rotacional na lei de Faraday, 

$$
\vec{\nabla}\times \vec{\nabla} \times \vec{E} = - \frac{\partial}{\partial t}(\vec{\nabla}\times \vec{B})
$$

e substituindo a Lei de Ampère e a Lei de Ohm, obtemos:

$$
\vec{\nabla} \times \vec{\nabla} \times \vec{E} = -\mu_0 \sigma \frac{\partial \vec{E}}{\partial t}
$$


Tomando, ainda, o divergente da Lei de Ampère, $\vec{\nabla}\cdot (\vec{\nabla}\times \vec{B}) = \mu_0\vec{\nabla}\cdot \vec{J}$, e usando a  identidade $\vec{\nabla}\cdot (\vec{\nabla}\times \vec{A}) = 0$, teremos

$$
\vec{\nabla} \cdot \vec{J} = 0
$$

Assumindo, ainda, que a condutividade é um constante $(\sigma_0)$, chegamos em

$$
\vec{\nabla}\cdot \vec{E} = 0
$$

De onde resulta, finalmente,

$$
\boxed{\vec{\nabla}^2 \vec{E} = \mu_0 \sigma_0 \frac{\partial \vec{E}}{\partial t}}
$$

Essas duas equações em $\vec{B}$ e $\vec{E}$:

$$
\boxed{\vec{\nabla}^2 \vec{E} = \mu_0 \sigma_0 \frac{\partial \vec{E}}{\partial t}} \hspace{0.8cm} \boxed{\vec{\nabla}^2 \vec{B} = \mu_0 \sigma \frac{\partial \vec{B}}{\partial t}}
$$

são as **equações de difusão**. 

Podemos, ainda, escrever essas equações na mesma forma da equação do Calor, 

$$
\boxed{ \frac{\partial \vec{E}}{\partial t} = \eta  \vec{\nabla}^2 \vec{E}} \hspace{0.8cm} \boxed{ \frac{\partial \vec{B}}{\partial t} = \eta \vec{\nabla}^2 \vec{B}}
$$

onde $\eta = 1/(\mu_0 \sigma_0)$ é a difusidade magnética, na qual depende da condutividade. A difusividade mede a rapidez com que o campo magnético se difunde através do material. Em um condutor perfeito ($\sigma \to \infty$), $\eta \to 0$, e o campo magnético fica "parado" no material. Em um mau condutor, $\eta$ é grande e o campo se dissipa rapidamente.

Para resolver essas equações, vamos aplicar separação de variáveis na equação:

$$
\frac{\partial \vec{F}}{\partial t} = \eta \vec{\nabla}^2 \vec{F}
$$

onde $\vec{F}$ pode ser $\vec{E}$ ou $\vec{B}$, componentes independentes entre si. Para cada componente, resolvemos:


$$
\frac{\partial}{\partial t}(F_x \hat{a}_x + F_j \hat{a}_j + F_z \hat{a}_z) = \eta \left(\left(\nabla^2 F_x\right) \hat{a}_x + \left(\nabla^2 F_j\right) \hat{a}_j + \left(\nabla^2 F_z\right) \hat{a}_z \right)
$$

Pela ortogonalidade dos vetores unitários, temos o conjunto de equações:

$$
\frac{\partial F_x}{\partial t} = \eta \nabla^2 F_x
$$
$$
\frac{\partial F_y}{\partial t} = \eta \nabla^2 F_y
$$
$$
\frac{\partial F_z}{\partial t} = \eta\nabla^2 F_z
$$

que são equações de difusão escalares. 

## Exemplo Aplicado


Consideremos uma placa condutora (por exemplo, de cobre ou um plasma) infinita nas direções $y$ e $z$, de espessura finita $L$ na direção $x$ ($0 < x < L$). O espaço fora da placa ($x < 0$ e $x > L$) é vácuo (isolante).

Como condição inicial ($t = 0$), assumimos um campo magnético constante $\vec{B}=B_0 \hat{a}_y$ dentro da placa $(0 < x < L)$. Para $t>0$, vamos assumir que o campo magnético externo é subitamente removido, de modo que o campo nas bordas da placa seja nulo: $B(0, t) = B(L, t) =0$. 

*O que queremos analisar é como o campo magnético irá decair e desaparecer com o tempo nesse condutor*. 

Em resumo, vamos resolver:


$$
\frac{\partial B}{\partial t} = \eta\frac{\partial ^2 B}{\partial x^2}, \hspace{0.4cm} 0< x < L, t > 0
$$
$$
B(0, t) = 0, \hspace{0.3cm} B(L, t) = 0
$$
$$
B(x, 0) = B_0
$$

Aqui, como o campo aponta na direção $\hat{a}_y$ e varia somente em $\hat{a}_x$, o campo resultante $\vec{B}$ também aponta na direção $\hat{a}_y$. Assim, por simplicidade, será representado apenas por $B(x,t)$ a magnitude da componente $\hat{a}_y$ do campo.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# espessura da placa
L = 1

fig, ax = plt.subplots(figsize=(6, 5))

# Região da placa condutora, ocupando 0 < x < L
placa = patches.Rectangle((0, -1.5), L, 3, facecolor='teal', edgecolor='black', label="Placa Condutora")
ax.add_patch(placa)

# Fronteiras da placa (linhas tracejadas)
ax.axvline(0, color='black', linestyle='--')
ax.axvline(L, color='black', linestyle='--')

# regiões
ax.text(-1.5, 0, 'Vácuo\n(Isolante)', ha='center', va='center', fontsize=10)
ax.text(L + 1.5, 0, 'Vácuo\n(Isolante)', ha='center', va='center', fontsize=10)
ax.text(L / 2, 0, 'Placa\nCondutora', ha='center', va='center', fontsize=10, color='white')

# Setas
ax.annotate('', xy=(0, 1.8), xytext=(L, 1.8), arrowprops=dict(arrowstyle='<->', color='black'))
ax.text(L / 2, 2.0, r'$L$', ha='center', fontsize=12)

# eixos
ax.set_xlim(-3, L + 3)
ax.set_ylim(-2.5, 2.5)
ax.set_xlabel('$x$', fontsize=12)
ax.set_yticks([])  # Remove y
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks([0, L])
ax.set_xticklabels(['$0$', '$L$'])
ax.grid(False)
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```

### Solução Analítica

Vamos assumir que a solução pode ser escrita pelo produto $B(x,t) = X(x)T(t)$. Substituindo isso na EDP:

$$
X(x)T'(t) = \eta T(t)X''(x) \Rightarrow \frac{1}{\eta}\frac{T'}{T} = \frac{X''}{X}
$$

Como o lado esquerdo depende apenas de $t$ e o valor direito apenas de $x$, essa igualdade será válida se ambos os lados forem iguais a uma constante. Nesse caso, vamos tomar a constante $-\lambda^2$. Assim, resulta duas equações diferenciais ordinárias:

- Equação Temporal:

$$
T' + \lambda^2 \eta T = 0
$$

- Equação Espacial: 

$$
X'' + \lambda^2 X = 0
$$

Para a equação temporal, teremos uma solução exponencial, dada por:

$$
r + \lambda^2 \eta = 0 \Rightarrow r = -\lambda^2 \eta
$$

$$
\Rightarrow \boxed{T(t) = Ce^{-\lambda^2 \eta t}}
$$

Para a equação espacial, teremos:

$$
r^2 + \lambda^2 = 0 \Rightarrow r = \pm i\lambda
$$

$$
\Rightarrow \boxed{X(x) = A\sin (\lambda x) + B\cos(\lambda x)}
$$

Aplicando as condições de contorno em $X(x)$:

$$
B(0, t) = 0 \Rightarrow X(0) = 0
$$

$$
A_1\sin(0) + A_2\cos (0) = 0 \Rightarrow \boxed{A_2 = 0}
$$

Portanto, a solução espacial se reduz a $X(x) = A_1\sin(\lambda x) = A\sin(\lambda x)$. 

$$
B(L, t) = 0 \Rightarrow X(L) = 0
$$

$$
A \sin (\lambda L) = 0
$$

Como estamos interessados na solução não trivial (coeficientes não identicamente nulos), devemos ter, pelo comportamento da função seno, 

$$
\lambda L = n\pi, \hspace{0.4cm} n = 1, 2, 3, \cdots
$$

$$
\Rightarrow \lambda_{n} = \frac{n\pi}{L}
$$

Portanto, 

$$
B_n(x,t) = X_n(x) T_n (t) = A_{n} \sin\left(\frac{n \pi x}{L} \right) e^{-\eta (n\pi/L)^2 t}
$$

Para solução completa,

$$
\boxed{B(x,t) = \sum_{n=1}^{\infty} A_{n} \sin\left(\frac{n \pi x}{L} \right) e^{-\eta (n\pi/L)^2 t}}
$$

Aqui, $A_n$ são os coeficientes da série a serem determinados. 

Para calcular os coeficientes, aplicamos a condição inicial:

$$
B(x,0) = B_0 \Rightarrow B_0 = \sum_{n=1}^{\infty} A_{n}\sin\left(\frac{n \pi x}{L} \right)
$$

Temos

$$
A_n = \frac{2}{L}\int_{0}^{L}B_0\sin\left(\frac{n \pi x}{L} \right) dx = \frac{2B_0}{L} \left[-\frac{L}{n\pi}\cos\left(\frac{n\pi x}{L} \right) \right]_{0}^{L}
$$

$$
A_{n} = - \frac{2B_0}{n\pi}(\cos (n\pi) - \cos(0)) = - \frac{2B_0}{n\pi}((-1)^n - 1)
$$

Pela paridade do valor do valor de $n$, temos:

- Se $n$ é par, $(-1)^n - 1 = 1 -1 = 0 \Rightarrow A_{n} =0$. 
- Se $n$ é ímpar, $(-1)^n -1 = -1 - 1 = -2$. 

Assim:

$$
\boxed{B(x,t) = \sum_{n = 1, 3, 5, \cdots}^{\infty} \frac{4B_0}{n\pi} \sin\left(\frac{n\pi x}{L} \right)e^{-\eta(n\pi/L)^2 t}}
$$

Abaixo segue os gráficos da solução analítica, usando como parâmetros $L=1$ (comprimento da placa), $\eta = 0.1$ (difusividade magnética) e $B_0 = 1$ (valor inicial de $B(x,0) = B_0$).

```python
# Gráfico da solução analítica

import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
L = 1.0       # comprimento da placa
eta = 0.1     # difusividade magnética
B0 = 1.0      # valor inicial de B(x, 0)

# Domínio espacial e tempos para análise
x = np.linspace(0, L, 300)
times = [0.000, 0.001, 0.01, 0.1, 0.5, 1.0]  # momentos no tempo
n_terms = 50  # número de termos na série de Fourier

# Função para calcular B(x, t)
def B_xt(x, t, L, eta, B0, n_terms):
    B = np.zeros_like(x)
    for n in range(1, 2 * n_terms, 2):  # apenas n ímpares
        coef = (4 * B0) / (n * np.pi)
        B += coef * np.sin(n * np.pi * x / L) * np.exp(-eta * (n * np.pi / L) ** 2 * t)
    return B

# Gráfico 2D
plt.style.use('bmh')
plt.figure(figsize=(8, 5))
for t in times:
    B = B_xt(x, t, L, eta, B0, n_terms)
    plt.plot(x, B, label=f't = {t:.3f}')

plt.xlabel('$x$')
plt.ylabel('$B(x, t)$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

As condições de contorno estão satisfeitas, isto é $x=0$ e $x=L$ para todo tempo - observado pelos extremos do gráfico.

A solução também respeita a dissipação do campo magnético, pelo decaimento exponencial dos termos da série, dado pelos fatores $e^{-\eta(n\pi/L)^2 t}$. 

```python
from mpl_toolkits.mplot3d import Axes3D

# Parâmetros de tempo e espaço 
t_vals = np.linspace(0, 1.0, 100)
x_vals = np.linspace(0, L, 300)
X, T = np.meshgrid(x_vals, t_vals)

# B(x,t) para cada ponto da malha
B_3d = np.array([B_xt(x_vals, t, L, eta, B0, n_terms) for t in t_vals])

# Gráfico 3D
plt.style.use('bmh')
fig3d = plt.figure(figsize=(10, 8))
ax3d = fig3d.add_subplot(111, projection='3d')
ax3d.plot_surface(X, T, B_3d, cmap='viridis')

ax3d.set_xlabel('$x$')
ax3d.set_ylabel('$t$')
ax3d.set_zlabel('$B(x,t)$')

plt.tight_layout()
plt.show()
```

### Solução Numérica

Para solução numérica, vamos aplicar a aproximação via diferenças finitas, pela fórmula:

\begin{align}
u_{i, j+1} &= u_{i, j} + \frac{kh_t}{h_x^2}\left(u_{i+1, j} - 2 u_{i, j} + u_{i-1, j}\right) \\ &= u_{i, j} + \lambda\left(u_{i+1, j} - 2 u_{i, j} + u_{i-1, j}\right),
\end{align}

onde

$$\lambda=\frac{k h_t}{h_x^2}$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
L = 1.0             # comprimento da barra
t_max = 1.0         # tempo total
eta = 0.1           # difusividade magnética

n = 50              # pontos espaciais
m = 1000            # pontos temporais

h_x = L / n
h_t = t_max / m
lambda_ = eta * h_t / h_x**2

# malhas 
xi = np.linspace(0, L, n+1)
tj = np.linspace(0, t_max, m+1)

# matriz de solução
u = np.zeros((len(xi), len(tj)))

# Condição inicial: B(x,0) = B0
B0 = 1.0
u[:, 0] = B0

# iterando o método de Gauss-Seidel
tol = 1e-4 # tolerância
err = tol + 1 # erro
while err > tol:
  u_ant = np.copy(u)
  # Itera-se no tempo 'j' para calcular o passo seguinte 'j+1'
  for j in range(0, len(tj) - 1):
    # Itera-se no espaço 'i'
    for i in range(1, len(xi) - 1):
      # Fórmula recursiva 
      u[i, j+1] = u[i, j] + lambda_ * (u[i+1, j] - 2*u[i, j] + u[i-1, j])
  
  # atualiza o erro
  err = np.linalg.norm(u - u_ant)    
    
    
# gráfico
plt.figure(figsize=(8, 5))
for k in [0, 10, 50, 200, 500, 1000]:
    plt.plot(xi, u[:, k], label=f't = {tj[k]:.3f}')
plt.xlabel('$x$')
plt.ylabel('$B(x,t)$')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

A linha para $t = 0.000$ representa a condição inicial de $B_0=1$ para todo $t$. Novamente, tem um decaimente do campo magnético conforme o tempo avança. 

```python
print(lambda_) # valor de lambda para averiguar a estabilidade
```

Aqui, $\lambda = 0.25$. O método anterior é estável e convergente para a EDP de interesse quando $\lambda < 0.25$. No caso acima, o método poderia ser instável, apresentando oscilações (o que não ocorreu) e também poderia divirgir (o que também não foi o caso).
