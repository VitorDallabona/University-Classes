**8.** O método de ortogonalização de Gram-Schmidt consiste em um algoritmo que permite obter um conjunto de vetores ortogonais entre si a partir de um conjunto linearmente independentes de outros vetores. Para um conjunto de três vetores $\{ v_1, v_2, v_3\}$ o processo pode ser descrito por
$$
\begin{aligned}
& u_1=v_1 \\

& u_3=v_3-\frac{\left\langle v_3, u_1\right\rangle}{\left\langle u_1, u_1\right\rangle} u_1-\frac{\left\langle v_3, u_2\right\rangle}{\left\langle u_2, u_2\right\rangle} u_2

**9.**  O parábola que passa pelos pontos $(x_i, y_i)$, $i = 1,2,3$, pode ser obtida resolvendo o sistema linear

$$ y_i = ax_i^2 + bx_i+c $$

Dados os pontos $(2,-2)$, $(3,-5)$ e $(4,0)$, escreva o sistema na forma matricial e resolva usando matriz inversa obtendo como solução os coeficientes $a$, $b$ e $c$. Então, plote o gráfico da parábola e dos pontos.

```python
# Pontos dados
pontos = np.array([[2, -2],
                   [3, -5],
                   [4, 0]])

# Criando a matriz de coeficientes (x^2, x, 1)
X = np.array([[ponto[0]**2, ponto[0], 1] for ponto in pontos])

# Criando o vetor de termos constantes (y)
y = pontos[:, 1]

# Calculando a matriz inversa de X
X_inversa = np.linalg.inv(X)

# Resolvendo o sistema linear para obter os coeficientes a, b e c
a, b, c = np.dot(X_inversa, y)

# Imprimindo os coeficientes
print("Coeficientes da parábola:")
print("a =", a)
print("b =", b)
print("c =", c)

# Criando pontos para plotar a parábola
x_parabola = np.linspace(1, 5, 100)
y_parabola = a * x_parabola**2 + b * x_parabola + c

# Plotando a parábola e os pontos dados
plt.plot(x_parabola, y_parabola, label='Parábola', color='black')
plt.scatter(pontos[:, 0], pontos[:, 1], color='red', label='Pontos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Parábola e dos Pontos')
plt.legend()
plt.grid(True)
plt.show()
```

**10.** Em muitas aplicações é necessário encontrar uma função para representar um conjunto de dados. Usando aproximação por *mínimos quadrados* podemos encontrar o polinômio de grau 3 que melhor representa um conjunto de 10 pontos $(x_i, y_i)$, $i=1,2,...,10$ resolvendo o sistema linear

$$(Z^TZ)a = Z^Ty$$

onde $Z$ é a matriz $10\times4$ obtida a partir dos valores de $x_i$, $i=1,2,...,10$ fazendo

$$Z = \left[\begin{matrix}
1 & x_1 & x_1^2 & x_1^3\\
1 & x_2 & x_2^2 & x_2^3\\
1 & x_3 & x_3^2 & x_3^3\\
\vdots & \vdots & \vdots & \vdots\\
1 & x_{10} & x_{10}^2 & x_{10}^3\\
\end{matrix}\right]$$

$a$ é a matriz dos coeficientes do polinômio $a=[a_0, a_1, a_2,a_3]^T$ e $y$ é dado por $y=[y_1,y_2,...,y_{10}]^T$.

Monte e resolva o sistema para encontrar o polinômio, então, plote o gráfico desse polinômio juntamente com os pontos:

x (m) |  1,3 |  0,9 |  2,2 | 1,95 | 1,78 | 3,1  | 2,48 | 2,74 | 3,35 | 3,99 |
------|------|------|------|------|------|------|------|------|------|------|
y (cm)|  1,9 | 3,01 | -3,1 | -1.98| -1.53| -5.01|-4.91 |-5.01 |-4.02 | 0,15 |

```python
points = np.array([[1.3 , 1.9], [0.9, 3.01], [2.2, -3.1], [1.95, -1.98], [1.78, -1.53],
                   [3.1, -5.01], [2.48, -4.91], [2.74, -5.01], [3.35, -4.02], [3.99, 0.15]])

Z = np.array([[1, ponto[0], ponto[0]**2, ponto[0]**3] for ponto in points]) # Definindo a matriz Z
print(Z)
```

```python
LS = np.transpose(Z)@Z # definindo a expressão Z^T * Z
print(LS)

y = np.array([[ponto[1]] for ponto in points]) # definindo a matriz y
print(np.transpose(y)) # matriz transposta de y
```

```python
RS = (Z.T)@y # lado direito da expressão
print(RS)
```

```python
np.linalg.solve(LS, RS) # resolvendo o sistema
```
